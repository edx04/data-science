```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/edx04/data-science/main/data/Employee.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
sns.FacetGrid(df,hue = 'City' , height = 5).map(plt.scatter,'Age','ExperienceInCurrentDomain').add_legend();
plt.title('Edad vs años de experiencia por ciudad', fontsize = 20)
plt.show()
```

```python
sns.set(font_scale = 1.5)
fig, ax = plt.subplots(1,1,figsize=(20,15))

plt.title("Año de incorporacion a la compañia vs años de experiencia")
g = sns.stripplot(x='JoiningYear', y='ExperienceInCurrentDomain', hue='LeaveOrNot', data=df,dodge=False,size=20,jitter=False,alpha=0.25,marker="d")

handles, labels = ax.get_legend_handles_labels()
labels = ["No renuncio","Renuncio"]
labels, handles = zip(*sorted(zip(labels, handles)))

ax.legend(handles, labels, title="Leave or not", bbox_to_anchor=(1.01, 1.01), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
q = sns.pairplot(df,hue = 'LeaveOrNot' , height = 3 , palette = 'Set2')
new_labels = ['Renuncio','Continua']
for t, l in zip(q._legend.texts, new_labels):
    t.set_text(l)
```

```python
sns.set(font_scale = 1)
p = sns.heatmap(df.corr(), cmap="YlGnBu", annot=True).set(title='Correlacion de variables')
```

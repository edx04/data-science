```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/edx04/data-science/main/data/Employee.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
sns.FacetGrid(df,hue = 'City' , size = 5).map(plt.scatter,'Age','ExperienceInCurrentDomain').add_legend();
plt.title('Edad vs a単os de experiencia por ciudad', fontsize = 20)
plt.show()
```

```python
sns.FacetGrid(df,hue = 'LeaveOrNot' , size = 5).map(plt.scatter,'JoiningYear','ExperienceInCurrentDomain').add_legend();
plt.title('A単o de incorporaciion a la compa単ia vs a単os de experiencia', fontsize = 20)
plt.show()
```

```python
sns.set(font_scale = 1.5)
sns.catplot(x='JoiningYear', y='ExperienceInCurrentDomain', hue='LeaveOrNot', kind="swarm", data=df).set(title='A単o de incorporaciion a la compa単ia vs a単os de experiencia')
```

```python
p = sns.pairplot(df,hue = 'LeaveOrNot' , size = 3 , palette = 'Set2')
new_labels = ['Renuncio','Continua']
for t, l in zip(p._legend.texts, new_labels):
    t.set_text(l)
```

```python
sns.heatmap(df.corr(), cmap="YlGnBu", annot=True).set(title='Correlacion de variables')
```

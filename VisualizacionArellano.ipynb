```python
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib as mpl
import seaborn as sns
```

```python
visualizacion = pd.read_csv('https://raw.githubusercontent.com/edx04/data-science/main/data/HRDataset_v14.csv')
```

```python
visualizacion.describe()
```

```python
salary = visualizacion['Salary']
sns.histplot(data=visualizacion, x ='Salary', fill=True )
```

```python
plt.figure(figsize=(25,12))
fig, ax = plt.subplots()
sns.catplot(data=visualizacion, kind='violin',x='GenderID',y='Salary')
```

```python
visualizacion.head(5)
```

```python
visualizacion['DateofHire'] = pd.to_datetime(visualizacion['DateofHire'] , format="%m/%d/%Y")
datos = visualizacion[['DateofHire','Employee_Name']].groupby('DateofHire').count()
datos = datos.rename(columns={'Employee_Name':'Contratados'})
```

```python
ax = datos.plot(figsize=(25,12),legend=False)
ax.set_ylabel('Contratados')
```
